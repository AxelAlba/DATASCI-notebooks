### Import packages

```python
from newsapi.newsapi_client import NewsApiClient
import pandas as pd
import json

api = NewsApiClient(api_key='9455267b4dbc4f8888adde0b98f023b4')
```

### News from CNN (March 11-12)

```python
results = api.get_everything(sources='cnn',
                                      from_param='2021-03-11',
                                      to='2021-03-12',
                                      language='en')
```

```python
results
```

### Convert the news gotten from the API to a dataframe & Implement Data Cleaning

```python
news = results["articles"]
news_df = pd.DataFrame([article for article in news])
news_df.head(10)
```

Removing columns: source, description, url, & urlToImage

```python
news_df = news_df.drop(['source', 'description', 'url', 'urlToImage'], axis=1)
news_df.head(3)
```

Removing rows that has a 'None' content

```python
news_df.dropna(inplace=True)
news_df
```

### Saving the dataframe into a JSON File

```python
#news_df.to_json(r'./CNN_March_11_12_data.json', orient="records") # No indents
json_file = news_df.to_json(orient="records")
json_news = json.loads(json_file)
news_formatted = json.dumps(json_news, indent=2)
with open("CNN_March_11_12_data.json", "w") as out:
    out.write(news_formatted)
```
